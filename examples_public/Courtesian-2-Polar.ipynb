# This example will show how to transform the image from Courtesian to Polar coordinates

```python
from maximus48.polar import polar_image
from maximus48 import polar2
import numpy as np
import tifffile
from maximus48 import var
import matplotlib.pyplot as plt
```

```python
#set the ROI of image first, the logic corresponds to FIJI (to be read (x,y,x1,y1 at the image - inverse to numpy!)
ROI = (20,20,2048,2048) 

# where to look for files
folder = '/Users/mpolikarpov/data_server/raw_data/tomo/'
image_name = 'Platy-Platy_9902_1_01800.tiff'

# import image
image = tifffile.imread(folder+image_name)[ROI[1]:ROI[3], ROI[0]:ROI[2]]

# calculate the image in polar coordinates
# by default, the origin is the center of the image 
# but you can specify the desired pixel passing "origin = (Numpy_coord_0, Numpy_coord_1)" to the function
out = polar2.reproject_image_into_polar(image)[:,:,0]
```

```python
# show original image 
plt.figure(num=None, figsize=(20, 20), facecolor='w', edgecolor='k')
plt.imshow(image, cmap='gray')
plt.title('original image', fontdict = {'fontsize' : 40})
plt.show()
```

```python
# show polar image 
plt.figure(num=None, figsize=(20, 20), facecolor='w', edgecolor='k')
plt.imshow(out, cmap='gray')
plt.title('polar image', fontdict = {'fontsize' : 40})
plt.show()
```
